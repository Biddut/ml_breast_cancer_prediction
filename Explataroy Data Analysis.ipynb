```python
#Insert your code here
for col_name in columns_list_mean:
  m = plt.hist(df[df["diagnosis"] == "M"].get(col_name),bins=30,fc = (1,0,0,0.5),label = "Malignant")
  b = plt.hist(df[df["diagnosis"] == "B"].get(col_name),bins=30,fc = (0,1,0,0.5),label = "Benign")
  plt.legend()
  plt.xlabel("%s values"%col_name)
  plt.ylabel("Frequency")
  plt.title("Histogram of %s for Benign and Malignant Tumors" %col_name)
  plt.show()
  print("Most frequent malignant %s is: %0.3f"%(col_name,m[1][list(m[0]).index(m[0].max())]))
  print("Most frequent Benign %s is: %0.3f"%(col_name,b[1][list(b[0]).index(b[0].max())]))
```
